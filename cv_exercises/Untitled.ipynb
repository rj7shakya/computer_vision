```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline 
```

```python
img = cv2.imread('birds.jpg')
img_copy = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_copy)
```

```python
smooth = cv2.GaussianBlur(img_copy,(9,9),0)
unsharp = cv2.addWeighted(img_copy,1.5,smooth,-0.5,0)
plt.imshow(unsharp)
```
