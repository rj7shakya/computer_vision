# color space in hsv


```python
#h:0-180,s:0-255,v:0-255
image = cv2.imread('red.jpg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow('image',hsv_image)
cv2.imshow('h',hsv_image[:,:,0])
cv2.imshow('s',hsv_image[:,:,1])
cv2.imshow('v',hsv_image[:,:,2])


cv2.waitKey()
cv2.destroyAllWindows()
```

# rgb channels

```python
import cv2

image = cv2.imread('red.jpg')
b,g,r = cv2.split(image)

print(b.shape)
cv2.imshow('r',r)
cv2.imshow('g',g)
cv2.imshow('b',b)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged = cv2.merge([b,g,r])
cv2.imshow('merged',merged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged = cv2.merge([b+100,g,r])
cv2.imshow('merged',merged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged = cv2.merge([b,g,r+100])
cv2.imshow('merged',merged)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np

image = cv2.imread('red.jpg')
b,g,r = cv2.split(image)
zeros = np.zeros(image.shape[:2],dtype='uint8')

cv2.imshow('red',cv2.merge([zeros,zeros,r]))
cv2.imshow('greeen',cv2.merge([zeros,g,zeros]))
cv2.imshow('blue',cv2.merge([b,zeros,zeros]))
cv2.waitKey()
cv2.destroyAllWindows()
```

# visualize individual color

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('red.jpg')
histogram = cv2.calcHist([image],[0],None,[256],[0,256])

plt.hist(image.ravel(),256,[0,256])
plt.show()

color =('b','g','r')

for i,col in enumerate(color):
    histogram2 = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram2,color = col)
    plt.xlim([0,256])
    
plt.show()
```

# drawing

```python
import cv2
import numpy as np

image = np.zeros((512,512,3),np.uint8)
cv2.line(image, (0,0) , (511,511) , (255,127,0) ,5)
cv2.imshow('blue line',image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#rectangle

image = np.zeros((512,512,3),np.uint8)
cv2.rectangle(image, (100,100) , (300,250) , (127,50,127) ,-1)
cv2.imshow('rectangle',image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#circle 

image = np.zeros((512,512,3),np.uint8)
cv2.circle(image, (350,350) ,100 , (255,127,0) ,-1)
cv2.imshow('circle',image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#polygon

image = np.zeros((512,512,3),np.uint8)

pts = np.array([[10,50],[400,50],[90,200],[50,200]],np.int32)
pts.reshape((-1,1,2))

cv2.polylines(image, [pts],True , (255,127,0) ,5)
cv2.imshow('polygon',image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#priting text
image = np.zeros((512,512,3),np.uint8)
cv2.putText(image,'Hello World',(75,290),cv2.FONT_HERSHEY_COMPLEX,2,(100,170,0),3)
cv2.imshow("hello",image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
